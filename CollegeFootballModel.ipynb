```python
import pandas
import sqlite3
import numpy
```

```python
conn = sqlite3.connect('/home/computerowner/Documents/PythonCodes/Bets/BettingData.db')
dfSeason = pandas.read_sql_query('SELECT * FROM SeasonResults', conn)
dfGame = pandas.read_sql_query('SELECT * FROM GameResults', conn)

dfSeason['PointsFor'] = pandas.to_numeric(dfSeason['PointsFor'])
```

```python
lastSeason = dfSeason.copy()
```

```python
uniqueSchoolNames = dfSeason.SchoolName.unique()
# print (uniqueSchoolNames)
uniqueYears = dfSeason.Year.unique()
# print (uniqueYears)
# uniqueYears.dtype
```

```python
yearList = list()
schoolList = list()
meanList = list()
medianList = list()
stdevList = list()
countList = list()

myDict = {}

for school in uniqueSchoolNames:
    for year in uniqueYears:
        thatYear = year
        yearList.append(year)
#         print (year)
        someSchool = lastSeason[(lastSeason['SchoolName'] == school) & (lastSeason['Year'] == year)].copy()
#         print (someSchool)
        schoolList.append(school)
#         print (school)
        mean = someSchool['PointsFor'].mean()
        meanList.append(mean)
#         print (mean)
        median = someSchool['PointsFor'].median()
        medianList.append(median)
#         print (median)
        stdev = someSchool['PointsFor'].std()
        stdevList.append(stdev)
#         print (stdev)
        count = len(someSchool.index)
        countList.append(count)
#         print (count)

myDict = {'year': yearList, 'schoolName':schoolList, 'mean':meanList, 'median':medianList, 'stdev':stdevList, 'count': countList} 
schoolStats = pandas.DataFrame(myDict)
schoolStats.dropna()
schoolStats
```
